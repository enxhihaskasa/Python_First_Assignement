```python
!pip install stanza

import stanza

# Download Albanian model (first time only)
stanza.download("sq")

# Load pipeline
nlp = stanza.Pipeline("sq")

# Load your raw Albanian text
with open("/content/01_01_AL_Pergatitjet_e_Pasqyrave_Financiare.txt", "r", encoding="utf-8") as f:
    text = f.read()

# Run NLP pipeline
doc = nlp(text)

# Print annotations
for sentence in doc.sentences:
    for word in sentence.words:
        print(f"{word.id} - {word.text} - {word.lemma} - {word.upos} - {word.deprel}")
    print()
```

```python
from transformers import pipeline

# Load multilingual BERT for masked word prediction
fill_mask = pipeline("fill-mask", model="bert-base-multilingual-cased")

# Try a sample sentence from your text (Albanian)
sample_text = "Raporti financiar është [MASK]."

# Run prediction
predictions = fill_mask(sample_text)

# Show top predictions
for pred in predictions:
    print(f">>> {pred['sequence']} (score: {pred['score']:.4f})")
```

```python
from transformers import pipeline

# Load a multilingual NER model
ner_pipe = pipeline("ner", model="Davlan/xlm-roberta-base-ner-hrl", aggregation_strategy="simple")

# Example Albanian sentence (or use a line from your file)
with open("/content/01_01_AL_Pergatitjet_e_Pasqyrave_Financiare.txt", "r", encoding="utf-8") as f:
    text = f.read()

# Run the NER pipeline
entities = ner_pipe(text)

# Print recognized entities
for entity in entities:
    print(f"{entity['word']} -> {entity['entity_group']} (score: {entity['score']:.2f})")
```
